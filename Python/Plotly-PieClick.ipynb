# Plotly Pie Click

This notebook demonstrates a quick interactive pie chart in which the index of the selected pie wedge is printed below the figure.

Source code from https://community.plot.ly/t/interactive-graph-click-event/19219

```python
import plotly.graph_objs as go
from ipywidgets import Output, VBox
```

```python
# create the Figure widget and use a graphics object to create the JSON for the pie chart
fig = go.FigureWidget([go.Pie(labels=['Alpha','Bravo','Charlie'], 
                              values=[1, 2, 3])])
pie = fig.data[0]

# Alternative way of creating the figue 
#fig = go.FigureWidget()
#pie = fig.add_pie(labels=['Alpha','Bravo','Charlie'],values=[1, 2, 3])

```

## Utilize ipywidgets.Output
Output widget captures and displays stdout and stderr as well as rich output generated by IPython.

https://ipywidgets.readthedocs.io/en/latest/examples/Output%20Widget.html

```python
# Capture the output.  
out = Output()
@out.capture(clear_output=True)
def handle_click(trace, points, state):
    print(points.point_inds[0])

# Setup the event handle for when the users clicks on the pie chart
pie.on_click(handle_click)
```

```python
# output the pie chart and output widget with a vertical container widget
VBox([fig, out])
```
